```python
null_columns = ['product_category_name', 'product_name_lenght', 'product_description_lenght', 'product_photos_qty',
                     'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']

products = products[null_columns].fillna(value='missing value')
```

```python

plt.figure(figsize=(10,7))
ax = sns.barplot(x=products['product_category_name'].value_counts().values, y=products['product_category_name'].value_counts().index, palette='Set2')
plt.title('Number of order per each product category')
max_values = products.groupby('product_category_name')['product_category_name'].count().max()
min_values = products.groupby('product_category_name')['product_category_name'].count().min()
for index, value in enumerate(products['product_category_name'].value_counts().values):
   if value == max_values:
       ax.text(value, index, f'{value}', ha='right', va='center')
   if value == min_values:
       ax.text(value, index, f'{value}', ha='left', va='center')
plt.show()
```

```python

def classify_product(x):
    categorys = {
    'Beauty & Health': ['health_beauty', 'perfumery', 'diapers_and_hygiene'],
    'Book & Stationary': ['stationery', 'books_general_interest', 'books_imported', 'books_technical'],
    'Electronics': ['computers_accessories', 'auto', 'air_conditioning', 'telephony', 'watches_gifts', 'consoles_games',
                    'electronics', 'small_appliances', 'small_appliances_home_oven_and_coffee', 'signaling_and_security',
                    'musical_instruments', 'fixed_telephony', 'tablets_printing_image', 'computers', 'audio',
                    'security_and_services'],
    'Entertainment': ['sports_leisure', 'toys', 'art', 'music', 'dvds_blu_ray', 'christmas_supplies', 'party_supplies',
                      'cine_photo', 'cds_dvds_musicals', 'arts_and_craftsmanship'],
    'Fashion': ['baby', 'fashion_female_clothing', 'cool_stuff', 'fashion_bags_accessories', 'fashion_male_clothing',
                'fashion_shoes', 'fashion_underwear_beach', 'fashion_sport', 'fashion_childrens_clothes'],
    'Food & Drinks': ['drinks', 'food'],
    'Furniture': ['office_furniture', 'home_confort', 'furniture_decor', 'bed_bath_table',
                  'kitchen_dining_laundry_garden_furniture', 'home_construction', 'furniture_living_room',
                  'furniture_bedroom', 'furniture_mattress_and_upholstery', 'home_comfort'],
    'Home & Garden': ['housewares', 'garden_tools', 'pet_shop', 'construction_tools_lights', 'luggage_accessories',
                      'home_appliances', 'home_appliances', 'market_place', 'costruction_tools_garden', 'la_cuisine',
                      'flowers'],
    'Industry & Construction': ['costruction_tools', 'construction_tools_construction',
                                'industry_commerce_and_business', 'construction_tools_safety',
                                'agro_industry_and_commerce']
}
    for category, keywords in categorys.items():
        if x in keywords:
            return category
    return 'Other'

products['product_category_name'] = products['product_category_name'].apply(classify_product)
```

```python

plt.figure(figsize=(10,7))
ax = sns.barplot(x=products['product_category_name'].value_counts().values, y=products['product_category_name'].value_counts().index, palette='Set2')
plt.title('Number of order per each product category')
max_values = products.groupby('product_category_name')['product_category_name'].count().max()
min_values = products.groupby('product_category_name')['product_category_name'].count().min()
for index, value in enumerate(products['product_category_name'].value_counts().values):
   if value == max_values:
       ax.text(value, index, f'{value}', ha='right', va='center')
   if value == min_values:
       ax.text(value, index, f'{value}', ha='left', va='center')
plt.show()
```
