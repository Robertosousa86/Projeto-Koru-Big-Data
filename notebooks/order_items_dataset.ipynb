3. Detectando e removendo possíveis Outliers.

```python
order_items[['price', 'freight_value']].describe().transpose()
```

Ao observar os valores da coluna "price", podemos notar que a diferença entre os 75% (Terceiro quartil) e o max é muito grande (134.90 vs 6735.00), o que indica que há pontos muito distantes da maioria dos dados. Essa disparidade sugere a presença de valores extremos ou outliers. Em contra partida os valores existentes na coluna "freight_value" podem indicar diversos fatores, como por exemplo: valor cobrado pela distancia do frete, peso do produto, promoções, impostos estaduais entre outros fatores que podem influenciar nos valores.

```python
plt.figure(figsize=(10,9))
sns.set_theme(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
sns.histplot(order_items['price'], bins=100)
plt.title('Distribution of data price')
plt.xlabel('Price')
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.boxplot(order_items['price'])
plt.title('Box Plot : Distribution of freight value')
plt.xlabel('Price')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.hist(order_items['price'], bins=30, edgecolor='black')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(dpi=150)
plt.figure(figsize=(14,7))
plt.subplot(212)
plt.plot(order_items['price'], 'x')
plt.title('Distribution of data price')
plt.ylabel('Price')
plt.xlabel('Count')
plt.grid()
plt.show()
```

Podemos observar nos gráficos acima que a escala do eixo x atinge 7000, gerando distorções. Isso ocorre porque há dados com valores próximos de 7000. Isso pode ser validado ao observar-mos a descrição estatística da coluna "price".

- Calculando os quartis

```python
Q1 = order_items['price'].quantile(0.25)
Q3 = order_items['price'].quantile(0.75)

print('Q1: %.2f' %Q1, '(Quartil inferior)')
print('Q3: %.2f' %Q3, '(Quartil superior)')
```

 Calculando intervalo interquartil:

- C -> Fator de escala que determina a extensão do intervalo interquartil usado para identificar outliers.
- IQR-> É a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1). Representa a dispersão central dos dados.
- IL (inferior limit) -> É o limite inferior para identificar outliers. Calculado subtraindo um múltiplo do IQR (C) do primeiro quartil (Q1).
- UL(upper limit) -> É o limite superior para identificar outliers. Calculado somando um múltiplo do IQR (C) ao terceiro quartil (Q3).

```python
C = 1.5

IQR = Q3 - Q1
IL = Q1 - C * IQR
UL = Q3 + C * IQR

print('Intervalo Interquartil: %.2f' %IQR)
print('Limites:')
print(' Limite inferior: %.2f' %IL)
print(' Limite superior: %.2f' %UL)
      
```

Removendo Outliers

```python
order_items = order_items[order_items['price'] < UL]
```

```python
plt.figure(figsize=(10,9))
sns.set_theme(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
sns.histplot(order_items['price'], bins=100)
plt.title('Distribution of data price')
plt.xlabel('Price')
plt.show()
```

```python
plt.figure(dpi=150)
plt.figure(figsize=(14,7))
plt.plot(order_items['price'], 'x')
plt.title('Distribution of data price')
plt.ylabel('Price')
plt.xlabel('Count')
plt.grid()
plt.show()
```
